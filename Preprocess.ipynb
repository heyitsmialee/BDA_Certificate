# Preprocess 예제 2 : Filtering & Sorting

```python
# Data Road
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
df
```

```python
df.loc[df['quantity']==3].head()
```

```python
df.loc[df['quantity']==3].head().reset_index(drop=True)
```

```python
# quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임 정의
df2 = df[['quantity', 'item_price']]
df2
```

```python
# item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장
df['new_price'] = df2['item_price'].str[1:].astype('float')
df
```

```python
# new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수
df3 = df.loc[df['new_price']<=5]
len(df3)
```

```python
# item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화
df4 = df.loc[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
df4
```

```python
# new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라
df5 = df.loc[(df['new_price']<=9)&(df['item_name']=='Chicken Salad Bowl')].head()
df5
```

```python
# df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라
df.sort_values('new_price', ascending=True).reset_index(drop=True)
```

```python
# df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라
df.loc[df['item_name'].str.contains('Chips')]
```

```python
# 짝수번째 컬럼만들 포함하는 데이터 프레임 출력
df.iloc[:,::2]
```

```python
# df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화
df.sort_values('new_price', ascending=False).reset_index(drop=True)
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱스
df.loc[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후
# item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

df6 = df.loc[(df['item_name']=='Steak Salad') | (df['item_name']=='Bowl')]
df6.drop_duplicates('item_name')
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후
# item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

df7 = df.loc[(df['item_name']=='Steak Salad') | (df['item_name']=='Bowl')]
df7.drop_duplicates('item_name', keep='last')
```

```python
# df 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터를 인덱싱
df.loc[df['new_price'] >= df['new_price'].mean()]
```

# Preprocess 3 : Grouping

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head()
```

```python
df.groupby('host_name').size()
```

```python
df.groupby('host_name').size().sort_values(ascending=False).to_frame().rename(columns={0:'counts'})
```

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
```

```python
df2 = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
df2.groupby(['neighbourhood_group'], as_index=False).max()
```

```python
# neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값
df3 = df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
df3
```

```python
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```
