# 1. 유튜브 인기동영상 데이터

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
df.info()
```

```python
# 인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)
df['channelTitle'].loc[df['channelTitle'].value_counts().sort_values(ascending=False).head(10)].unique()
```

```python
answer =list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
print(answer)
```

```python
df.info()
```

```python
df['channelTitle'].loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index)].unique()
```

```python
# 논란으로 인기동영상이 된 케이스를 확인하고 싶다.
# dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라

df['channelTitle'].loc[df['likes'] < df['dislikes']].unique()
```

```python
# 채널명을 바꾼 케이스가 있는지 확인하고 싶다.
# channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라
change = df[['channelTitle', 'channelId']].drop_duplicates().channelId.value_counts()
change = change[change>1]
len(change)
```

```python
# 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?

df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df2 = df.loc[df['trending_date2'].dt.weekday == 6]
df2['categoryId'].value_counts().head(1).index[0]
```

```python
# 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
df2 = df[['trending_date2', 'categoryId']].groupby([df['trending_date2'].dt.day_name(), 'categoryId'],as_index=False).size()
df2.pivot_table(index='categoryId', columns='trending_date2')
```

```python
# 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.
# viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)

df2 = df.copy()
df2 = df2.loc[df2['view_count']!=0]

df2['ratio'] = df2['comment_count']/df['view_count']
df['title'].iloc[df2['ratio'].sort_values(ascending=False).head(1).index].unique()
```

```python
# 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.
# viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)

df2 = df.copy()
df2 = df2.loc[df['view_count'] !=0]
df2['ratio'] = df2['comment_count']/df2['view_count']
df2 = df2.loc[df2['ratio'] !=0]
df2.sort_values(by='ratio').iloc[0].title
```

```python
# like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)

tg = df.copy()
tg = tg.loc[(tg['likes']!=0) & (tg['dislikes']!=0)]

tg['ratio'] = tg['dislikes']/tg['likes']
tg.sort_values(by='ratio').iloc[0].title
```

```python
# 가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜기준, 중복포함)
# 채널아이디 = 채널아이디의 최댓값의 아이디 인 타이틀을 가지고 옴
df['channelTitle'].loc[df['channelId'] == df.groupby('channelId').size().sort_values(ascending=False).index[0]].unique()
```

```python
# 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?

(df[['title','channelId']].value_counts()>=20).sum()
```

2. 유튜브 공범컨텐츠 동영상 데이터

```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
# 각 데이터의 ‘ct’컬럼을 시간으로 인식할수 있게 datatype을 변경하고
# video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라

channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])

video.groupby('videoname').size()
```

```python
# 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라
video[['videoname','ct','viewcnt']].sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')
```

```python
channel.info()
```

```python
# Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

df = channel[['ct','channelname','subcnt']].loc[channel['ct'] >= pd.to_datetime('2021-10-03')].sort_values(['channelname','ct']).drop_duplicates('channelname', keep='first')
df.iloc[:,1:].reset_index(drop=True)
```

```python
# 각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라
channel.info()

start = channel.loc[channel['ct'].dt.strftime('%Y-%m-%d %H') == '2021-10-03 03']
end = channel.loc[channel['ct'].dt.strftime('%Y-%m-%d %H') == '2021-11-01 15']

start_df = start[['channelname','subcnt']].reset_index(drop=True)
end_df = end[['channelname','subcnt']].reset_index(drop=True)

start_df.columns = ['channelname', 'start']
end_df.columns = ['channelname', 'end']

final_df = pd.merge(start_df, end_df)
final_df['del'] = final_df['end'] - final_df['start']
final_df[['channelname', 'del']]
```

```python
# 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다.
# 공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라
import datetime

display(video.head())

df2 = video.loc[video['videoname'].str.contains('1')].sort_values('ct').reset_index(drop=True)

df2[(df2.ct.diff(1) >= datetime.timedelta(minutes=20)) | (df2.ct.diff(1) <= datetime.timedelta(minutes=5))]

df2[df2.index.isin([720, 721, 722, 723, 1635, 1636, 1637])]
```

```python
# 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라
df2 = video.sort_values(['ct', 'videoname']).drop_duplicates('videoname')
df2['date'] = df2['ct'].dt.date
df2[['videoname','date']]
```

```python
# “공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.
# 공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

video['time'] = video.ct.dt.hour

video[['viewcnt','ct', 'videoname']].loc[video['time'] == 21].sort_values(['viewcnt','ct']).drop_duplicates('videoname').sort_values('viewcnt', ascending=False).reset_index(drop=True)
```

```python
# video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고
# videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

df = video.sort_values(['videoname','ct']).drop_duplicates('videoname', keep='last')
df['ratio'] = df['dislikecnt'] / df['likecnt']
df[['videoname','ratio']].sort_values('ratio').reset_index(drop=True)
```

```python
# 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

start = pd.to_datetime('2021-11-01 00')
end = pd.to_datetime('2021-11-01 15')

df = video.loc[(video['ct'] >= start) & (video['ct'] <= end)].reset_index(drop=True)
def check(x):
  result = max(x) - min(x)
  return result

df[['videoname', 'viewcnt']].groupby('videoname').agg(check)
```

```python
# video 데이터 중에서 중복되는 데이터가 존재한다. 중복되는 각 데이터의 시간대와 videoname 을 구하여라

duplicated = video[video.index.isin(set(video.index) - set(video.drop_duplicates().index))]
duplicated[['videoname', 'ct']]
```
